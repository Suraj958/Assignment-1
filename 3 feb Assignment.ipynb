```python
# Q2
""" *args' and '*kwargs' allows us to pass a variable number of non-keyword arguments to a Python function.
'*args' is used to pass a non-keyworded, variable-length argument  to the function. 
It allows you to pass an arbitrary number of arguments as a list.
'**kwargs' is used to pass keyworded, variable-length arguments to the function. 
It allows you to pass an arbitrary number of keyword arguments as a dict """

print('function for *args')
def print_args(*args):
    print(args)
print_args(1,2,3,4,5,6)

print('function for **kwargs')

def print_kwargs(**kwargs):
    print(kwargs)

print_kwargs(name='suraj', age=21, city='lucknow')

```

```python
# Q4
''' A generator-function is defined like a normal function, but whenever it needs to generate a value,
it does so with the yield keyword rather than return.
The yield keyword is used in a generator function to return a value and pause the function's execution.'''
def cube(n):
    for i in range(1, n+1):
        yield i**3

# Initialize the generator object
generator = cube(10)

# Iterate through the values generated by the generator function
for i in generator:
    print(i)
```

```python
# Q5
def prime_numbers(limit):
    for number in range(2, limit):
        for i in range(2, int(number**0.5)+1):
            if number % i == 0:
                break
        else:
            yield number

# Initialize the generator object
prime_gen = prime_numbers(1000)

# Print the first 20 prime numbers
for i in range(20):
    print(next(prime_gen))
```

```python
# Q8
n = int(input("Enter a number: "))
temp = n
reverse = 0
while n > 0:
    rem = n % 10
    reverse = reverse * 10 + rem
    n = n // 10
if temp == reverse:
    print(temp, "is a Palindrome")
else:
    print(temp, "is not a Palindrome")
```
